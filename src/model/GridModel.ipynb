```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import math
from sklearn import metrics
```

```python
df = pd.read_csv("../../dataset/grid.csv")
df.head()
```

```python
df["Duration"] = (df["Timestamp2"] - df["Timestamp1"])*60
df.head() 
```

```python
indexNames  =  df[df["Duration"]<=0].index
df = df.drop(indexNames, inplace = False)
```

```python
df["Distance"] = df["Distance"] * 1000 #Distance in meters
df.head()
```

```python
df["Speed"] = df["Distance"] / df["Duration"] #in m/s
df.head()
```

```python
df.describe()
```

```python
columns = ["LAT1", "LON1", "Timestamp1", "LAT2", "LON2", "Angle", "East", "West", "North", "South"]
X = df[columns]
X.describe()
```

```python
y = df["Speed"]
y.head()
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
model = LinearRegression()
model.fit(train_X, train_y)
preds = model.predict(val_X)
np.sqrt(metrics.mean_squared_error(val_y, preds))
```

```python
def Haversine(lat1,lon1,lat2,lon2):
    R = 6371.0088
    lat1,lon1,lat2,lon2 = map(np.radians, [lat1,lon1,lat2,lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2) **2
    c = 2 * np.arctan2(a**0.5, (1-a)**0.5)
    d = R * c
    return round(d,4)
```

```python
def angleFromCoordinate(LAT1, LON1, LAT2, LON2):
    dLon = (LON2 - LON1)
    y = np.sin(dLon) * np.cos(LAT2)
    x = np.cos(LAT1) * np.sin(LAT2) - np.sin(LAT1) * np.cos(LAT2) * np.cos(dLon)
    brng = np.arctan2(y, x)
    brng = np.degrees(brng)
    brng = (brng + 360) % 360
    brng = 360 - brng
    return brng
```

```python
distance = Haversine(df.loc[0, "LAT1"], df.loc[0, "LON1"], df.loc[6, "LAT1"], df.loc[6, "LON1"])
distance = distance*1000
print(distance)
```

```python
angle = angleFromCoordinate(df.loc[0, "LAT1"], df.loc[0, "LON1"], df.loc[6, "LAT1"], df.loc[6, "LON1"])
print(angle)
```
